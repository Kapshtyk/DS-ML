# Gold recovery

The main objective of this project is to build models that predict gold recovery rates from gold-bearing ore at different stages of the production process: for the rough feed and for the final concentrate. 

We have data with mining and cleaning parameters split into a training sample and a test sample. An initial data set has also been provided. Some parameters cannot be used for model training because they are calculated at a later stage of the process. Because of this, the test sample lacks some attributes that may be in the training sample. Also there are no target features in the test sample.

A number of additional tasks also had to be solved:
- to check the correctness of the calculation of the ore recovery efficiency (our target trait) - it was necessary to do an independent calculation using a formula, and to calculate the MAE metric between the data from the dataset and also the results of our calculation;
- to study the change in metal concentrations at different cleaning stages, to investigate the total concentration of elements at different stages of the production process, and to compare the distribution of some of the features in the training and test samples;
- determine the function to be used as a metric to score the quality of the models.

We will look at the gaps in more detail, for which we will define the function.

```python
def missing_values_table(df):
        # Number of missing values
        mis_val = df.isna().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isna().sum() / len(df)
        
        # Creating a dataframe
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Renaming the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing values', 1 : '% of total values'})
        
        # Filtering and sorting the resulting dataframe
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of total values', ascending=False).round(1)
        
        return mis_val_table_ren_columns
```

The spread of values in almost all the columns in which we will fill in the blanks is quite large. Given the order of the numbers, we could immediately fill in the gaps in the `rougher.input.floatbank11_xanthate` column with the median value, but first we will try to use a method that will predict a more accurate value. Given the continuity of the production process (we have data for 24 hours of each day), we can assume that the values should vary less during a given day than when analysing changes in the indicator over a longer period of time. Let's try to fill in the gaps with this assumption in mind. 

We have a unique attribute for each of the records stored in the `date` column, which contains a string representation of the data in `datetime` format. For our task we don't need to convert this data to the appropriate format yet, we just extract the date in string format from this column.

```python
data_full['date_str'] = data_full['date'].apply(lambda x: x[0:10:1])
```

Next, we will identify the dates on which we had gaps in the columns from the `columns_with_na` list. 

```python
dates_with_na = []
for i in columns_with_na:
    dates_with_na += list(data_full[data_full[i].isna()]['date_str'].unique())
```

Next, we will create a dictionary whose keys will be the rows with the date on which there was an omission in any of the values of interest, and whose value will be a nested dictionary in which the keys will be the names of the four columns of interest, and the values will be the median value of the relevant column for the corresponding day.

```python
for_missing_values = {}
for i in dates_with_na:
        _slice = data_full[data_full['date_str'] == i]
        for_missing_values[i] = {
       'rougher.state.floatbank10_e_air': _slice['rougher.state.floatbank10_e_air'].median(),
       'rougher.input.floatbank11_xanthate': _slice['rougher.input.floatbank11_xanthate'].median(),
       'secondary_cleaner.state.floatbank2_a_air': _slice['secondary_cleaner.state.floatbank2_a_air'].median(),
       'rougher.input.feed_size' : _slice['rougher.input.feed_size'].median()}
```

Define a function that will take as its argument the dataset and the name of the column in which we are going to fill in the gaps. The function will be applied to the row of the dataset. If the date in string format specified in the `date_str` column is not in the `dates_with_na` list, the function returns the current value from the column whose gaps we are trying to fill. Otherwise, the function returns the median of the value sought for the relevant day.   

```python
def na_filler(data_full, column):
    if data_full['date_str'] not in dates_with_na:
        return data_full[column]
    else:
        return(for_missing_values[data_full['date_str']][column])
```

Apply the function in a cycle, keeping in mind the number of skips before using it in order to assess the result of the function.

```python
before = missing_values_table(data_full[test_columns])
before
```

```python
for i in columns_with_na:
    data_full[i] = data_full.apply(na_filler, axis=1, column=i)
```

```python
after = missing_values_table(data_full[test_columns])
after
```

```python
print('Initially there were {} gaps in the dataset, after calling the function the number of gaps was {}, we filled in {} ({:.2%}) gaps'
      .format(before['Missing values'].sum(), after['Missing values'].sum(),
       before['Missing values'].sum() - after['Missing values'].sum(),
       (before['Missing values'].sum() - after['Missing values'].sum()) / 
        before ['Missing values'].sum()))
```

We got rid of a third of the gaps, which is already good. We fill in the rest of the gaps using the attribute ffill and limit=12 (half a day), and fill in the remaining gaps with the median value over the whole dataset.

```python
for i in columns_with_na:
    data_full[i].fillna(method='ffill', limit=12, inplace=True)
missing_values_table(data_full[test_columns])
```

```python
for i in columns_with_na:
    data_full[i].fillna(data_full[i].median(), inplace=True)
missing_values_table(data_full[test_columns])
```

### Verifying that the calculation of the target feature is correct

Next, we will check if the target attribute `rougher.output.recovery` is calculated correctly by creating a function that will take as its argument a dataset, after which it will calculate enrichment efficiency for each row using the given formula. To avoid exceptions, we will add a try/except construct, in the second part of which we will append the `indexes` list with the date in string format for the row in which the efficiency calculation failed.

```python
indexes = []
def recovery(data_full):
    F = data_full['rougher.input.feed_au' ]
    C = data_full['rougher.output.concentrate_au'] 
    T = data_full['rougher.output.tail_au'] 
    try:
        return (C * (F - T)) / (F * (C - T)) * 100
    except:
        indexes.append(data_full['date_str'])
        pass
```

```python
data_full['recovery_raw_my'] = data_full.apply(recovery, axis=1)
recovery_test = data_full[['date'] + ['recovery_raw_my'] + ['rougher.output.recovery']]
recovery_test.head()
```

```python
recovery_test.isna().value_counts()
```

```python
print(indexes)
```

Given the work we did to remove outliers from the target attribute columns, we added a column to the `data_full` dataset containing the recovery efficiency calculation. There were no problems with the calculation (except in the function never worked, so the list of `indexes` is empty). Let's look at the mean absolute error for the specified target feature.

```python
mae_rougher = mean_absolute_error(data_full['recovery_raw_my'], data_full['rougher.output.recovery'])
mae_rougher
```

The mean absolute error for the target feature `rougher.output.recovery' is negligible, it can be assumed that the calculation is correct.

### Создание кастомной метрики для модели

Напишем функцию для рассчета симметричного среднего абсолютного процентного отклонения, а также с помощью метода `make_scorer()` создадим оценщика для наших моделей. Так как нас интересуют модели с минимальным показателем smape, мы будем использовать атрибут `greater_is_better=False`.

```python
def smape(targets, predictions):
    return abs(1 / len(targets) * np.sum(abs(targets - predictions) / 
                                       ((abs(targets) + abs(predictions)) / 2)) * 100)

scorer = make_scorer(smape, greater_is_better=False)
```

Сразу же напишем функцию, которая позволит осуществлять рассчет итогового sMAPE, чтобы комплексно посмотреть на метрики полученных моделей.

```python
def final_smape(rougher_test_features, rougher_test_targets, rougher_model, 
                final_test_features, final_test_targets, final_model):
    rougher_score = smape(rougher_model.predict(rougher_test_features), rougher_test_targets)
    final_score = smape(final_model.predict(final_test_features), final_test_targets)
    return 0.25 * rougher_score + 0.75 * final_score
```

### Подбор моделей и их гиперпараметров

Определим функцию, которая будет осуществлять перебор нескольких моделей в цикле, и, в зависимости от вида модели, будет осуществлять GridSearch в соответствии переданной для каждого доступного вида модели сеткой параметров (предусмотрено значение по умолчанию). Оценка модели будет осуществляться с помощью созданного выше оценщика, основанного на функции smape. Название лучшей модели, её оценка и гиперпараметры будут возвращаться функцией. В случае, если модель не сможет показать лучший результат, чем Dummy Regressor, будет возвращено соответствующее сообщение.

```python
def modeller(model, features, target, baseline, 
             lr_param=[{'normalize' : [True, False]}],
             rfr_param=[{'n_estimators': range(1, 51, 10), 
                         'max_depth': range(1, 7, 1), 
                         'min_samples_leaf': range(1, 4), 
                         'min_samples_split': range(2, 4, 2)}],
             knn_param=[{'n_neighbors' : range(3, 140, 2),
                         'weights' : ['distance', 'uniform']}],
             metrics='neg_mean_absolute_error'):
    
    best_score = float('inf')
    best_model = None
    best_params = None
    
    for i in model:
        if type(i) == LinearRegression:
            param_grid_lr=lr_param
            grid = GridSearchCV(i, param_grid=param_grid_lr, scoring=metrics, verbose=4)
            grid.fit(features, target)
            score = grid.score(features, target)
            if abs(score) < best_score:
                best_score = abs(score)
                best_model = i
                best_params = grid.best_params_
        
        elif type(i) == RandomForestRegressor:
            param_grid_rfr=rfr_param
            grid = GridSearchCV(i, param_grid=param_grid_rfr, scoring=metrics, verbose=4)
            grid.fit(features, target)
            score = grid.score(features, target)
            if abs(score) < best_score:
                best_score = abs(score)
                best_model = i
                best_params = grid.best_params_
        
        elif type(i) == KNeighborsRegressor:
            param_grid_knn=knn_param
            grid = GridSearchCV(i, param_grid=param_grid_knn, scoring=metrics, verbose=4)
            grid.fit(features, target)
            score = grid.score(features, target)
            if abs(score) < best_score:
                best_score = abs(score)
                best_model = i
                best_params = grid.best_params_
    
    if best_score > baseline:
        print('Ни одна из моделей не показала лучший результат, чем DummyRegressor')
    
    else:
        return best_score, best_model, best_params
```

Создадим экземпляры классов `LinearRegression()` и `RandomForestRegressor()`, для последнего передадим гиперпараметр `random_state=12345`. Поместим созданные экземпляры в список, чтобы передать его в качестве атрибута в функцию `modeller()`.

```python
lr = LinearRegression()
rfr = RandomForestRegressor(random_state=12345)
models = [lr, rfr]
```

Мы выбрали две модели, для которых осуществили подбор параметров. Лучшую оценку показала модель RandomForestRegressor с указанными выше гиперпараметрами. Значение smape для нее на обучающей выборке составило 2.86%. Создадим и обучим модель с найденными гиперпараметрами, чтобы можно было протестировать её на тестовой выборке. 

```python
final_model_rougher = RandomForestRegressor(random_state=12345, 
                                            max_depth=8, 
                                            min_samples_leaf=2, 
                                            min_samples_split=2,
                                            n_estimators=41)
final_model_rougher.fit(rougher_train_features, rougher_train_targets)
predictions = final_model_rougher.predict(rougher_test_features)
final_model_rougher_score = smape(rougher_test_targets, predictions)
final_model_rougher_score
```

На тестовых данных наша модель достаточно качественно предсказала значение целевого признака `rougher.output.recovery`, показатель smape составил 4.30%.
